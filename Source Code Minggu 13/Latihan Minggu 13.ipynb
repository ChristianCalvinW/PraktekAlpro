```python
# Latihan 12.2
l = ["Calvin", "Calvin", "Ndus", "Daren", "Nicho"]
s = set(l)
print(f"List {l} menjadi set = {s}")
print("Set menghilangkan Duplikat dalam List")
print()

print(f"Set {s} menjadi list = ", end="")
s = list(s)
print(s)
print("Set menjadi list tidak ada perubahan pada value, hanya tanda kurung yang digunakan")
print()

t = ("Calvin", "Alven", "Ndus", "Darren", "Calvin")
s1 = set(t)
print(f"Tuple {t} menjadi Set = {s1}")
print("Sekali lagi Set menghilangkan value duplikat yang ada dalam tuple sebelumnya")
print()

print(f"Set {s1} menjadi Tuple = ", end="")
s1 = tuple(s1)
print(s1)
print("Sama seperti sebelumnya, Set menjadi Tuple hanya berubah pada tanda kurung")
```
